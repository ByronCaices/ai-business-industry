## 2. Variables y Tipos de Datos Básicos

Las **variables** son como cajas con nombre donde guardamos información. Python reconoce automáticamente el tipo de dato.

```python
# Variables numéricas enteras (int)
cantidad_productos = 150
empleados_turno = 12

# Variables numéricas decimales (float)
precio_unitario = 29.99
tasa_defectos = 0.03  # 3% de defectos

# Variables de texto (str)
producto = "Tornillo M8"
planta = "Planta Santiago Norte"

# Variables booleanas (True/False)
disponible = True
meta_cumplida = False

# Veamos qué guardamos
print("Producto:", producto)
print("Cantidad:", cantidad_productos)
print("Precio:", precio_unitario)
print("Disponible:", disponible)
print()

# Podemos ver el tipo de cada variable
print("Tipo de 'cantidad_productos':", type(cantidad_productos))
print("Tipo de 'precio_unitario':", type(precio_unitario))
print("Tipo de 'producto':", type(producto))
print("Tipo de 'disponible':", type(disponible))
```

## 3. Operaciones Matemáticas Básicas

Python funciona como una calculadora muy poderosa. Veamos operaciones típicas de gestión industrial.

```python
# Datos de producción
precio = 2500        # precio por unidad en pesos
cantidad = 200       # unidades producidas
costo_material = 800 # costo de material por unidad

# Operaciones básicas
costo_total = precio * cantidad
print("Ingreso total por ventas: $", costo_total)

ganancia_unitaria = precio - costo_material
print("Ganancia por unidad: $", ganancia_unitaria)

ganancia_total = ganancia_unitaria * cantidad
print("Ganancia total: $", ganancia_total)

# Producción promedio diaria (5 días de trabajo)
produccion_semanal = 750
promedio_diario = produccion_semanal / 5
print("Producción promedio diaria:", promedio_diario, "unidades")

# Potencias y módulo
print("\n--- Operaciones extra ---")
print("2 elevado a 10:", 2 ** 10)
print("750 unidades repartidas en lotes de 12:", 750 // 12, "lotes completos")
print("Unidades sobrantes:", 750 % 12)
```

## 4. Trabajando con Listas

Las **listas** permiten guardar múltiples datos en una sola variable. Son ideales para manejar datos de producción, inventario, etc.

```python
# Producción diaria de una semana (lunes a viernes)
produccion_diaria = [120, 135, 98, 145, 110]
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"]

print("Producción de la semana:", produccion_diaria)
print("Producción del lunes:", produccion_diaria[0])    # primer elemento (índice 0)
print("Producción del viernes:", produccion_diaria[4])   # último elemento
print("Producción del viernes:", produccion_diaria[-1])  # también el último (truco útil)

# Agregar el dato del sábado
produccion_diaria.append(50)
print("\nCon sábado incluido:", produccion_diaria)

# Estadísticas básicas con listas
print("\n--- Estadísticas de la semana ---")
print("Total producido:", sum(produccion_diaria), "unidades")
print("Días registrados:", len(produccion_diaria))
print("Mejor día:", max(produccion_diaria), "unidades")
print("Peor día:", min(produccion_diaria), "unidades")
print("Promedio diario:", round(sum(produccion_diaria) / len(produccion_diaria), 1), "unidades")
```

## 5. Condicionales: Tomar Decisiones con `if` / `else`

Los condicionales permiten que el programa **tome decisiones** según condiciones. Muy útil para control de calidad y alertas.

```python
# Clasificar nivel de producción
produccion_hoy = 135
meta_diaria = 130

if produccion_hoy >= meta_diaria:
    print(f"¡Meta cumplida! Se produjeron {produccion_hoy} unidades (meta: {meta_diaria})")
else:
    diferencia = meta_diaria - produccion_hoy
    print(f"Meta NO alcanzada. Faltan {diferencia} unidades.")

# Clasificar con múltiples niveles usando elif
print("\n--- Clasificación de rendimiento ---")
produccion = 145

if produccion >= 150:
    nivel = "EXCELENTE"
elif produccion >= 130:
    nivel = "BUENO"
elif produccion >= 100:
    nivel = "REGULAR"
else:
    nivel = "BAJO"

print(f"Producción: {produccion} unidades → Nivel: {nivel}")

# Control de calidad
print("\n--- Control de calidad ---")
defectos = 8
total_piezas = 200
tasa = defectos / total_piezas * 100

if tasa <= 2:
    print(f"Tasa de defectos: {tasa}% → APROBADO ✓")
else:
    print(f"Tasa de defectos: {tasa}% → REQUIERE REVISIÓN ✗")
```

## 6. Bucles: Repetir Acciones con `for`

Los **bucles** permiten repetir una acción para cada elemento de una lista. Muy útil para procesar datos de producción.

```python
# Recorrer la producción de cada día
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"]
produccion = [120, 135, 98, 145, 110]

print("=== Reporte Diario de Producción ===")
for i in range(len(dias)):
    print(f"  {dias[i]}: {produccion[i]} unidades")

# Acumulador: sumar el total con un bucle
print("\n=== Cálculo de producción total ===")
total = 0
for unidades in produccion:
    total = total + unidades
    print(f"  Sumando {unidades}... Total parcial: {total}")

print(f"\nProducción total de la semana: {total} unidades")

# Clasificar cada día automáticamente
print("\n=== Clasificación automática ===")
meta = 120
for i in range(len(dias)):
    estado = "✓ Cumple" if produccion[i] >= meta else "✗ No cumple"
    print(f"  {dias[i]}: {produccion[i]} uds → {estado}")
```

## 7. Funciones: Reutilizar Código

Las **funciones** permiten empaquetar código que usamos frecuentemente. Se definen con `def` y se llaman por su nombre.

```python
# Función para calcular costo total
def calcular_costo(precio, cantidad):
    """Calcula el costo total de un pedido."""
    total = precio * cantidad
    return total

# Usemos la función varias veces
print("Pedido 1:", calcular_costo(2500, 100), "pesos")
print("Pedido 2:", calcular_costo(1800, 250), "pesos")
print("Pedido 3:", calcular_costo(5200, 50), "pesos")

# Función para control de calidad
def control_calidad(defectos, total_piezas, umbral=2.0):
    """Determina si un lote pasa control de calidad."""
    tasa = defectos / total_piezas * 100
    if tasa <= umbral:
        return f"APROBADO (tasa: {tasa:.1f}%)"
    else:
        return f"RECHAZADO (tasa: {tasa:.1f}%, máximo permitido: {umbral}%)"

# Evaluar varios lotes
print("\n=== Control de Calidad ===")
print("Lote A:", control_calidad(3, 200))
print("Lote B:", control_calidad(12, 200))
print("Lote C:", control_calidad(5, 500))
```

```python
# Importar las librerías que usaremos
import matplotlib.pyplot as plt  # Para crear gráficos
import numpy as np               # Para cálculos numéricos
import pandas as pd              # Para manejar tablas de datos

print("¡Librerías cargadas correctamente!")
print("- matplotlib: para crear gráficos y visualizaciones")
print("- numpy: para cálculos numéricos eficientes")
print("- pandas: para manejar datos en tablas (como Excel)")
```

## 8. Tu Primer Gráfico: Producción Semanal (Líneas)

Un gráfico de líneas es perfecto para ver **tendencias** a lo largo del tiempo.

```python
# Datos de producción semanal
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"]
produccion = [120, 135, 98, 145, 110]
meta = 120  # meta diaria

# Crear el gráfico de líneas
plt.figure(figsize=(10, 5))
plt.plot(dias, produccion, marker='o', color='#00a499', linewidth=2, markersize=8, label='Producción')
plt.axhline(y=meta, color='#EA7600', linestyle='--', linewidth=1.5, label=f'Meta diaria ({meta} uds)')

# Personalizar
plt.title("Producción Diaria - Semana 1", fontsize=14, fontweight='bold')
plt.xlabel("Día de la semana")
plt.ylabel("Unidades producidas")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 9. Gráfico de Barras: Producción por Turno

Los gráficos de barras son ideales para **comparar categorías**.

```python
# Producción por turno
turnos = ["Mañana", "Tarde", "Noche"]
produccion_turnos = [450, 380, 290]
colores = ['#00a499', '#EA7600', '#5B5EA6']

plt.figure(figsize=(8, 5))
barras = plt.bar(turnos, produccion_turnos, color=colores, edgecolor='white', linewidth=1.5)

# Agregar valores encima de cada barra
for barra, valor in zip(barras, produccion_turnos):
    plt.text(barra.get_x() + barra.get_width()/2, barra.get_height() + 5,
             str(valor), ha='center', fontweight='bold', fontsize=12)

plt.title("Producción por Turno - Planta Santiago", fontsize=14, fontweight='bold')
plt.ylabel("Unidades producidas")
plt.ylim(0, 520)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 10. Gráfico de Torta: Distribución de Defectos

Los gráficos circulares muestran la **composición** o proporción de un total.

```python
# Tipos de defectos detectados en producción
tipos_defecto = ["Dimensional", "Superficial", "Funcional", "Otros"]
cantidades = [45, 30, 15, 10]
colores = ['#00a499', '#EA7600', '#5B5EA6', '#95C8D8']
explode = (0.08, 0, 0, 0)  # Destacar el defecto más frecuente

plt.figure(figsize=(8, 6))
plt.pie(cantidades, labels=tipos_defecto, autopct='%1.1f%%',
        colors=colores, explode=explode, shadow=True, startangle=90,
        textprops={'fontsize': 12})
plt.title("Distribución de Defectos en Producción", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 12. Análisis de Datos con Pandas

**Pandas** permite trabajar con datos en tablas (como Excel pero mucho más poderoso). Vamos a crear una tabla de datos de producción.

```python
# Crear una tabla (DataFrame) con datos de producción
datos = {
    "Día": ["Lunes", "Lunes", "Martes", "Martes", "Miércoles",
            "Miércoles", "Jueves", "Jueves", "Viernes", "Viernes"],
    "Turno": ["Mañana", "Tarde"] * 5,
    "Unidades": [120, 105, 135, 98, 142, 110, 128, 115, 145, 130],
    "Defectos": [3, 5, 2, 8, 1, 4, 3, 6, 2, 3]
}

df = pd.DataFrame(datos)

# Mostrar la tabla completa
print("=== Tabla de Producción Semanal ===")
print(df)
print()

# Estadísticas descriptivas automáticas
print("=== Estadísticas Descriptivas ===")
print(df.describe())
print()

# Filtrar: ¿Qué turnos tuvieron más de 4 defectos?
print("=== Turnos con más de 4 defectos ===")
problematicos = df[df["Defectos"] > 4]
print(problematicos)
```

## 13. Visualización Directa desde Pandas

Pandas permite crear gráficos **directamente desde una tabla** con una sola línea de código.

```python
# Gráfico de barras agrupadas: producción por día y turno
resumen = df.pivot_table(values="Unidades", index="Día", columns="Turno", aggfunc="sum")

# Reordenar los días de la semana
orden_dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"]
resumen = resumen.reindex(orden_dias)

resumen.plot(kind='bar', figsize=(10, 5), color=['#00a499', '#EA7600'], edgecolor='white')
plt.title("Producción por Día y Turno", fontsize=14, fontweight='bold')
plt.ylabel("Unidades producidas")
plt.xlabel("")
plt.xticks(rotation=0)
plt.legend(title="Turno")
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Gráfico de línea: tendencia de defectos
print()
df_defectos = df.groupby("Día")["Defectos"].sum().reindex(orden_dias)
df_defectos.plot(kind='line', marker='o', color='#EA7600', figsize=(10, 4), linewidth=2)
plt.title("Total de Defectos por Día", fontsize=14, fontweight='bold')
plt.ylabel("Cantidad de defectos")
plt.xlabel("")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 14. Mini Ejercicio Integrador: Análisis de Producción

Ejercicio guiado: analicemos datos de producción de **4 semanas**. Completen las partes marcadas con `# TODO`.

```python
# Datos de producción: 4 semanas (20 días hábiles)
produccion_mes = [
    120, 135, 98, 145, 110,   # Semana 1
    125, 140, 105, 150, 115,  # Semana 2
    130, 128, 142, 138, 120,  # Semana 3
    145, 150, 135, 155, 140   # Semana 4
]

# Calcular promedios semanales
semanas = [produccion_mes[i:i+5] for i in range(0, 20, 5)]
promedios = [sum(s) / len(s) for s in semanas]

print("=== Promedios Semanales ===")
for i, prom in enumerate(promedios):
    print(f"  Semana {i+1}: {prom:.1f} unidades/día")

# Mejor y peor día del mes
mejor = max(produccion_mes)
peor = min(produccion_mes)
print(f"\nMejor día: {mejor} unidades (día {produccion_mes.index(mejor) + 1})")
print(f"Peor día: {peor} unidades (día {produccion_mes.index(peor) + 1})")

# Gráfico combinado: barras + línea de promedio móvil
dias_mes = list(range(1, 21))

# Promedio móvil de 3 días
promedio_movil = []
for i in range(len(produccion_mes)):
    if i < 2:
        promedio_movil.append(sum(produccion_mes[:i+1]) / (i+1))
    else:
        promedio_movil.append(sum(produccion_mes[i-2:i+1]) / 3)

plt.figure(figsize=(12, 5))
plt.bar(dias_mes, produccion_mes, color='#00a499', alpha=0.6, label='Producción diaria')
plt.plot(dias_mes, promedio_movil, color='#EA7600', linewidth=2.5, marker='', label='Promedio móvil (3 días)')
plt.axhline(y=sum(produccion_mes)/len(produccion_mes), color='gray', linestyle=':', label='Promedio mensual')

# TODO: Prueben cambiar los colores del gráfico
# TODO: Prueben agregar una línea de meta en 140 unidades con plt.axhline()

plt.title("Producción del Mes con Promedio Móvil", fontsize=14, fontweight='bold')
plt.xlabel("Día del mes")
plt.ylabel("Unidades producidas")
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.xticks(dias_mes)
plt.tight_layout()
plt.show()
```

## 15. Un Paso Más: Regresión Lineal Simple

La **regresión lineal** es uno de los modelos más simples de Machine Learning. Encuentra la **mejor línea recta** que se ajusta a los datos para hacer predicciones.

Ejemplo: ¿Podemos predecir el **precio de una propiedad** según sus metros cuadrados?

```python
from sklearn.linear_model import LinearRegression

# Datos: metros cuadrados vs precio (en millones de pesos)
metros = np.array([40, 55, 60, 72, 85, 95, 110, 120, 135, 150]).reshape(-1, 1)
precio = np.array([45, 58, 62, 78, 90, 102, 118, 128, 142, 160])

# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(metros, precio)

# Predecir
metros_pred = np.linspace(30, 160, 100).reshape(-1, 1)
precio_pred = modelo.predict(metros_pred)

# Hacer una predicción específica
nuevo_depto = 80  # metros cuadrados
precio_estimado = modelo.predict([[nuevo_depto]])[0]

# Graficar
plt.figure(figsize=(10, 6))
plt.scatter(metros, precio, color='#00a499', s=100, zorder=5, label='Datos reales')
plt.plot(metros_pred, precio_pred, color='#EA7600', linewidth=2, label='Línea de regresión')
plt.scatter([nuevo_depto], [precio_estimado], color='red', s=150, marker='*', zorder=6,
            label=f'Predicción: {nuevo_depto}m² → ${precio_estimado:.1f}M')

plt.title("Regresión Lineal: Precio vs Metros Cuadrados", fontsize=14, fontweight='bold')
plt.xlabel("Metros cuadrados")
plt.ylabel("Precio (millones $CLP)")
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n¿Cuánto costaría un departamento de {nuevo_depto} m²?")
print(f"Precio estimado: ${precio_estimado:.1f} millones de pesos")
print(f"\nEl modelo aprendió que por cada m² extra, el precio sube ~${modelo.coef_[0]:.2f} millones")
```
