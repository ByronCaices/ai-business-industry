## 🧠 Introducción

Una parte esencial del ciclo de vida de un modelo de IA es su evaluación. Dependiendo del tipo de problema (clasificación o regresión), las métricas cambian:

- Clasificación: *accuracy*, *precision*, *recall*, *f1-score*
- Regresión: *MAE*, *MSE*, *R²*

Este laboratorio aplicará estas métricas a modelos previamente entrenados sobre datos industriales simulados.



```python
# Importar librerías y métricas necesarias
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix, classification_report,
                             mean_absolute_error, mean_squared_error, r2_score)
from sklearn.linear_model import LogisticRegression, LinearRegression

# Dataset simulado para clasificación binaria
np.random.seed(0)
X_class = np.random.rand(500, 5)
y_class = (X_class[:, 0] + X_class[:, 1] > 1).astype(int)

# Separación de datos en entrenamiento y prueba
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_class, y_class, test_size=0.2)

# Entrenamiento del modelo de clasificación
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train_c, y_train_c)
y_pred_c = clf.predict(X_test_c)

# Evaluación del modelo de clasificación
print("📊 Evaluación de Clasificación:")
print("Accuracy:", accuracy_score(y_test_c, y_pred_c))
print("Precision:", precision_score(y_test_c, y_pred_c))
print("Recall:", recall_score(y_test_c, y_pred_c))
print("F1 Score:", f1_score(y_test_c, y_pred_c))
print("Matriz de Confusión:", confusion_matrix(y_test_c, y_pred_c))
print("Reporte de Clasificación:", classification_report(y_test_c, y_pred_c))
```

```python
# Dataset simulado para regresión
X_reg = np.random.rand(500, 3)
y_reg = 3*X_reg[:, 0] - 2*X_reg[:, 1] + X_reg[:, 2] + np.random.randn(500)*0.2

# División de datos en entrenamiento y prueba
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_reg, y_reg, test_size=0.2)

# Entrenamiento del modelo de regresión lineal
reg = LinearRegression()
reg.fit(X_train_r, y_train_r)
y_pred_r = reg.predict(X_test_r)

# Evaluación del modelo de regresión
print("📈 Evaluación de Regresión:")
print("MAE:", mean_absolute_error(y_test_r, y_pred_r))
print("MSE:", mean_squared_error(y_test_r, y_pred_r))
print("R² Score:", r2_score(y_test_r, y_pred_r))
```
