## 🧠 Introducción al Aprendizaje Supervisado
El aprendizaje supervisado consiste en entrenar un modelo a partir de un conjunto de datos etiquetado, es decir, donde conocemos las respuestas correctas. El objetivo es que el modelo aprenda a generalizar para hacer predicciones sobre nuevos datos no vistos.

- En clasificación, las etiquetas son categorías.
- En regresión, las etiquetas son valores numéricos continuos.

## 🌳 Clasificación con Árbol de Decisión

```python
# Importar librerías necesarias
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

# Cargar dataset Iris
iris = load_iris()
X = iris.data  # características: medidas de las flores
y = iris.target  # etiquetas: tipo de flor

# Separar datos en entrenamiento y prueba (70%/30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar modelo de árbol de decisión
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Hacer predicciones con el conjunto de prueba
y_pred = clf.predict(X_test)

# Evaluar desempeño del modelo
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred))

# Visualizar el árbol resultante
plt.figure(figsize=(12,6))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Árbol de Decisión - Clasificación Iris")
plt.show()
```

## 📈 Regresión Lineal

```python
# Importar librerías para regresión
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.datasets import fetch_california_housing

# Dataset: California Housing
housing = fetch_california_housing()
X = housing.data  # variables predictoras
y = housing.target  # precio medio de las casas

# División de datos (entrenamiento y prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelo de regresión lineal
reg = LinearRegression()
reg.fit(X_train, y_train)

# Predicción y evaluación
y_pred = reg.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Error cuadrático medio (MSE):", mse)
```

## 📊 Interpretación de Resultados
**Clasificación:** Un modelo con alta precisión indica que está clasificando correctamente las flores. El árbol de decisión también es interpretable visualmente.

**Regresión:** El MSE nos da una idea de cuán lejos están las predicciones de los valores reales. Mientras más bajo, mejor.
