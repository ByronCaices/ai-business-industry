## 🧠 Mini Proyecto – Clasificación multiclase de fallas en sistemas industriales

```python
# Simulación de datos multiclase
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)  # semilla para reproducibilidad
N = 1200  # número total de registros simulados
X0 = np.random.normal(loc=[2, 2], scale=0.8, size=(N//3, 2))  # fallas tipo 0
X1 = np.random.normal(loc=[6, 2], scale=0.8, size=(N//3, 2))  # fallas tipo 1
X2 = np.random.normal(loc=[4, 6], scale=0.8, size=(N//3, 2))  # fallas tipo 2
X = np.vstack([X0, X1, X2])
y = np.array([0]*(N//3) + [1]*(N//3) + [2]*(N//3))  # etiquetas de clase

data = pd.DataFrame(X, columns=['presion', 'temperatura'])  # datos en tabla
data['falla'] = y
data.head()
```

```python
# Visualización de datos
sns.scatterplot(x='presion', y='temperatura', hue='falla', palette='deep', data=data)  # gráfico dispersión
plt.title('Tipos de falla industrial simulada');
```

```python
# Entrenamiento con MLP (Red Neuronal Multicapa)
import os, warnings
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'  # evita mensajes de TensorFlow
os.environ['CUDA_VISIBLE_DEVICES'] = '-1'  # fuerza uso de CPU
warnings.filterwarnings('ignore')  # suprime advertencias
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

X = data[['presion', 'temperatura']].values  # variables de entrada
y = to_categorical(data['falla'].values, num_classes=3)  # salida en one-hot

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)  # dividir datos

model = Sequential([
    Dense(16, activation='relu', input_shape=(2,)),  # capa de entrada
    Dense(12, activation='relu'),  # capa oculta
    Dense(3, activation='softmax')  # capa de salida con 3 clases
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])  # configurar entrenamiento
history = model.fit(X_train, y_train, epochs=40, validation_split=0.2, verbose=0)  # entrenar modelo
```

```python
# Evaluación del modelo
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f'Precisión en test: {accuracy:.4f}')  # rendimiento final
```
