## 🧠 Introducción

**Transfer learning** o transferencia de aprendizaje permite reutilizar redes ya entrenadas para tareas similares. Esto es muy útil en la industria, donde puede que no tengamos millones de datos.

Utilizaremos un modelo base (`MobileNetV2`) para clasificar imágenes simuladas de productos.



```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import warnings  # noqa: E402
import numpy as np  # noqa: E402
import tensorflow as tf  # noqa: E402
from tensorflow.keras.applications import MobileNetV2  # noqa: E402
from tensorflow.keras.models import Model  # noqa: E402
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D  # noqa: E402
from tensorflow.keras.preprocessing import image  # noqa: E402

warnings.filterwarnings('ignore')  # oculta advertencias

# Generador que normaliza imágenes y separa un 20% para validación
img_gen = image.ImageDataGenerator(rescale=1.0/255, validation_split=0.2)

# Datos ficticios de productos industriales (10 imágenes de 128x128)
dummy_images = np.random.rand(10, 128, 128, 3)
dummy_labels = tf.keras.utils.to_categorical(np.random.randint(2, size=10), 2)
data_flow = img_gen.flow(dummy_images, dummy_labels, batch_size=2)
batch_imgs, batch_lbls = next(data_flow)
print('Lote simulado:', batch_imgs.shape)

# Cargamos MobileNetV2 preentrenado y congelamos sus pesos
base_model = MobileNetV2(
    weights='imagenet', include_top=False, input_shape=(128, 128, 3)
 )
base_model.trainable = False  # mantenemos aprendizaje original

# Capas finales para clasificar nuestros productos (2 categorías)
x = base_model.output
x = GlobalAveragePooling2D()(x)  # reduce mapas a vector
x = Dense(64, activation='relu')(x)  # capa intermedia
predictions = Dense(2, activation='softmax')(x)  # dos probabilidades

model = Model(inputs=base_model.input, outputs=predictions)
model.compile(
    optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']
 )

# Entrenamos una vez con el lote simulado para verificar el flujo
model.train_on_batch(batch_imgs, batch_lbls)
print('Modelo de transferencia preparado.')
```
