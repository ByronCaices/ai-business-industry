## 🧠 Introducción

Una IA responsable debe considerar no solo la precisión del modelo, sino también:

- Transparencia
- Seguridad
- No discriminación
- Impacto social y ambiental

En este laboratorio construiremos un “checklist de responsabilidad” y aplicaremos simulaciones de riesgo en modelos de IA usados en producción.



```python
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Importamos librerías y fijamos semilla
np.random.seed(42)  # reproducibilidad

# Generamos datos sintéticos del proceso industrial
X = pd.DataFrame({
    'velocidad': np.random.normal(50, 10, 500),
    'presion': np.random.normal(5, 1, 500),
    'temperatura': np.random.normal(200, 15, 500),
    'operario_experto': np.random.choice([0, 1], size=500, p=[0.4, 0.6]),
})
# Regla simple para identificar defectos
y = ((X['velocidad'] > 55) & (X['temperatura'] > 210)).astype(int)

# Separamos datos para entrenar y probar
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Creamos y entrenamos el modelo
model = GradientBoostingClassifier()
model.fit(X_train, y_train)

# Realizamos predicciones y evaluamos precisión
y_pred = model.predict(X_test)
print("Precisión del modelo:", accuracy_score(y_test, y_pred))
```


## 🔎 Evaluación de Riesgos y Responsabilidad

A continuación, se presentan posibles consideraciones para una IA Responsable:

- ¿El modelo ha sido validado con distintos subconjuntos?
- ¿Contiene variables sensibles o indirectamente sesgadas?
- ¿Es transparente (interpretable)?
- ¿Se monitorea su rendimiento en producción?
- ¿Qué sucede si el modelo falla?
