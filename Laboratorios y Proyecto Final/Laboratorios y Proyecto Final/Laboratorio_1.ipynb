## 🧮 Fundamentos de Python

### Variables y tipos de datos

```python
# Definimos una variable de texto (string) con el nombre de la planta
nombre = "Planta Santiago"  # Esto es un texto, va entre comillas

# Definimos una variable numérica entera (int) para la producción diaria
produccion_diaria = 1580  # Número entero, sin decimales

# Definimos una variable numérica entera para la cantidad de fallas en el día
fallas_dia = 3  # Número entero, representa la cantidad de fallas

# Mostramos un mensaje en pantalla usando las variables anteriores
print(f"Hoy en {nombre} hubo {fallas_dia} fallas")  # La 'f' permite insertar variables dentro del texto
```

### Operaciones básicas

```python
# Calculamos la producción semanal multiplicando la producción diaria por 7 días
total_semana = produccion_diaria * 7  # Multiplicación de variables numéricas

# Mostramos el resultado en pantalla
print("Producción semanal estimada:", total_semana)  # La coma separa el texto del valor de la variable
```

### Comentarios y buenas prácticas

```python
# Calculamos el porcentaje de fallas del día
porcentaje_fallas = (fallas_dia / produccion_diaria) * 100  # División y multiplicación para obtener porcentaje

# Mostramos el porcentaje en pantalla
print("% de fallas hoy:", porcentaje_fallas)  # Imprime el texto y el valor calculado
```

## 📦 Uso de librerías básicas

```python
# Importamos la librería numpy para trabajar con números y arreglos
import numpy as np  # 'np' es un nombre corto para usar la librería más fácilmente

# Importamos la librería pandas para trabajar con tablas de datos
import pandas as pd  # 'pd' es un nombre corto para pandas

# Mostramos un mensaje para confirmar que las librerías se cargaron bien
print("Librerías cargadas correctamente")
```

## 📊 Visualización básica de datos

```python
# Importamos la librería para gráficos
import matplotlib.pyplot as plt  # 'plt' es el nombre corto para usar matplotlib

# Creamos una lista con la producción de cada día de la semana
produccion = [1520, 1600, 1580, 1595, 1610, 1570, 1588]  # Lista de números

# Dibujamos un gráfico de línea con los datos de producción
plt.plot(produccion)  # Dibuja la línea

# Agregamos título y nombres a los ejes
plt.title("Producción diaria")  # Título del gráfico
plt.xlabel("Día")  # Nombre del eje X
plt.ylabel("Unidades")  # Nombre del eje Y

# Mostramos el gráfico en pantalla
plt.show()
```
