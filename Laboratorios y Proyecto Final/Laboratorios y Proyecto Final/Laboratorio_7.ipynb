## 🧠 Introducción

A medida que los problemas industriales se vuelven más complejos, es necesario utilizar redes más profundas que puedan capturar relaciones no lineales entre los datos.

En este laboratorio entrenaremos una red neuronal profunda (deep neural network) usando Keras, con múltiples capas ocultas y funciones de activación ReLU.

Aplicaremos esta arquitectura a un caso de predicción industrial (rendimiento de máquinas).



```python
# Importar librerías necesarias
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'  # suprimir avisos de TensorFlow
os.environ['CUDA_VISIBLE_DEVICES'] = '-1'  # desactivar GPU para evitar mensajes
import numpy as np
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Simulación de datos de rendimiento industrial
np.random.seed(0)  # semilla para reproducibilidad
X = np.random.rand(1000, 10)  # características simuladas
y = 5*X[:,0] - 3*X[:,1] + 2*X[:,2] + np.random.randn(1000)*0.5  # salida con ruido

# División y normalización de los datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Definir la red neuronal profunda
model = keras.models.Sequential([
    keras.layers.Input(shape=(X_train.shape[1],)),  # capa de entrada
    keras.layers.Dense(64, activation='relu'),  # primera capa oculta
    keras.layers.Dense(64, activation='relu'),  # segunda capa oculta
    keras.layers.Dense(1)  # capa de salida para regresión
])

# Configurar el entrenamiento
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Entrenar el modelo
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=0)
```

```python
# Evaluar el modelo con datos de prueba
loss, mae = model.evaluate(X_test, y_test, verbose=0)
print(f"Error absoluto medio (MAE): {mae:.2f}")
```

```python
# Graficar la evolución del error durante el entrenamiento
plt.plot(history.history['loss'], label='Pérdida entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida validación')
plt.xlabel('Épocas')
plt.ylabel('MSE')
plt.legend()
plt.title("Curva de entrenamiento")
plt.grid(True)
plt.show()
```
