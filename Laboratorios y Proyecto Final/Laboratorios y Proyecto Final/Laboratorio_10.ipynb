## 🧠 Introducción

Uno de los desafíos de los modelos de IA es su carácter de "caja negra". Las técnicas de **interpretabilidad** nos ayudan a comprender por qué un modelo toma ciertas decisiones.

En este laboratorio exploraremos:

- Importancia de características
- Visualización de pesos
- Interpretación local con LIME

**Aplicación:** clasificación de fallas industriales simuladas



```python
# Importamos librerías necesarias
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Generamos datos simulados representando mediciones industriales
np.random.seed(0)
X = pd.DataFrame({
    'temperatura': np.random.normal(70, 5, 500)
    , 'vibracion': np.random.normal(1.2, 0.3, 500)
    , 'presion': np.random.normal(30, 4, 500)
    , 'flujo': np.random.normal(100, 15, 500)
    , 'corriente': np.random.normal(10, 2, 500)
})
y = ((X['temperatura'] > 72) & (X['presion'] > 32)).astype(int)

# Separación de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Entrenamos un bosque aleatorio para clasificar fallas
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)

# Evaluamos el modelo con un reporte de clasificación
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Visualizamos la relevancia de cada variable
importances = model.feature_importances_
features = X.columns

plt.figure(figsize=(8,5))
sns.barplot(x=importances, y=features)
plt.title("Importancia de Características")
plt.xlabel("Importancia")
plt.grid(True)
plt.show()
```

```python
# Interpretación local con LIME
# Ejecuta la siguiente línea solo si la librería no está instalada
# !pip install lime --quiet --disable-pip-version-check
from lime.lime_tabular import LimeTabularExplainer
from IPython.display import display, HTML

# Creamos el explicador con los datos de entrenamiento
explainer = LimeTabularExplainer(
    training_data=np.array(X_train),
    feature_names=X.columns,
    class_names=["Normal", "Falla"],
    mode="classification"
)

# Función auxiliar para evitar advertencias al predecir
def predict_proba_wrapper(x):
    x_df = pd.DataFrame(x, columns=X.columns)
    return model.predict_proba(x_df)

# Elegimos una muestra y generamos la explicación
i = 10  # índice de la observación a explicar
exp = explainer.explain_instance(X_test.iloc[i].values, predict_proba_wrapper)
display(HTML(exp.as_html()))
```
