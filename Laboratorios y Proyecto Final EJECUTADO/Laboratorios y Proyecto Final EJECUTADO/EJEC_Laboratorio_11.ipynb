## 🧠 Introducción

La implementación de IA en entornos industriales plantea nuevos desafíos éticos y normativos:

- Sesgos en modelos
- Privacidad de datos
- Responsabilidad ante errores
- Regulaciones emergentes

Este laboratorio incluye un ejercicio práctico de detección de sesgos y análisis de impactos.



```python
import numpy as np  # cálculos numéricos
import pandas as pd  # manejo de tablas
from sklearn.linear_model import LogisticRegression  # modelo simple de clasificación
from sklearn.metrics import classification_report  # métricas de desempeño
from sklearn.model_selection import train_test_split  # separar datos en train/test

# Generamos datos ficticios de postulantes
np.random.seed(42)
n = 500  # cantidad de casos
genero = np.random.choice(['masculino', 'femenino'], size=n, p=[0.6, 0.4])
experiencia = np.random.normal(5, 2, size=n)  # años de experiencia
puntaje = experiencia * 1.2 + (genero == 'masculino') * 1.5 + np.random.randn(n)
contratado = (puntaje > np.percentile(puntaje, 60)).astype(int)  # decisión final

df = pd.DataFrame({'genero': genero, 'experiencia': experiencia, 'contratado': contratado})  # tabla completa

# Codificamos 'genero' a número para usarlo en el modelo
df['genero_cod'] = (df['genero'] == 'masculino').astype(int)
X = df[['experiencia', 'genero_cod']]
y = df['contratado']

# Dividimos datos para entrenar y evaluar
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = LogisticRegression(max_iter=1000, random_state=42)  # ajustamos el modelo
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Medimos el desempeño del modelo
print('Evaluación del modelo con atributo sensible (género):')
print(classification_report(y_test, y_pred))
```
