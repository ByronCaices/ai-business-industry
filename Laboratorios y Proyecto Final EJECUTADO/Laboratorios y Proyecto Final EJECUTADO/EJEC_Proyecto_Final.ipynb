## 🧪 Caso de ejemplo opcional (para quienes no tengan datos)

Simulación de control de calidad en productos industriales:

- Variables: temperatura, vibración, presión, velocidad, operador
- Objetivo: clasificar productos como aceptables (0) o defectuosos (1)

Este dataset puede generarse usando NumPy si no se cuenta con datos reales.



```python

# Simulación de datos (puede usarse como base del proyecto)
import numpy as np
import pandas as pd

np.random.seed(42)
n = 600
data = pd.DataFrame({
    'temperatura': np.random.normal(70, 5, n),
    'vibracion': np.random.normal(1.1, 0.2, n),
    'presion': np.random.normal(32, 3, n),
    'velocidad': np.random.normal(45, 8, n),
    'operador': np.random.choice(['A', 'B', 'C'], size=n),
})

# Variable objetivo (calidad)
data['defectuoso'] = ((data['temperatura'] > 73) & (data['presion'] > 34)).astype(int)

data.head()
```
