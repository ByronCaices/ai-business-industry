## 🧠 Introducción teórica

En este laboratorio abordaremos dos aplicaciones industriales concretas:

- **Clasificación de fallas en motores industriales** utilizando árboles de decisión
- **Predicción del rendimiento de producción** mediante regresión lineal

Se utilizarán datasets simulados que imitan el comportamiento de sistemas industriales reales.

---


```python
# Importar librerías y simular datos de sensores
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generar mediciones simuladas
np.random.seed(42)
n = 500
X_motor = pd.DataFrame({
    'temperatura': np.random.normal(70, 10, n),
    'vibracion': np.random.normal(5, 1.5, n),
    'presion': np.random.normal(30, 5, n),
    'horas_uso': np.random.normal(1000, 300, n)
})
X_motor['falla'] = ((X_motor['temperatura'] > 75) &
                    (X_motor['vibracion'] > 6) &
                    (X_motor['presion'] > 32)).astype(int)

# Separar variable objetivo y dividir datos
y_motor = X_motor.pop('falla')
X_train, X_test, y_train, y_test = train_test_split(X_motor, y_motor, test_size=0.3, random_state=42)

# Entrenar árbol de decisión y evaluar
clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"Precisión (accuracy) en validación: {acc:.2f}")
```

```python
# Graficar árbol de decisión y relevancia de variables
plt.figure(figsize=(12, 6))
plot_tree(clf, feature_names=X_motor.columns, class_names=["No Falla", "Falla"], filled=True)
plt.title("Árbol de decisión - Fallas en motores")
plt.show()

# Mostrar importancia de cada atributo
importancias = clf.feature_importances_
sns.barplot(x=importancias, y=X_motor.columns)
plt.title("Importancia de variables")
plt.show()
```

```python
# Simular datos sobre productividad en planta
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X_prod = pd.DataFrame({
    'velocidad_maquina': np.random.normal(50, 5, n),
    'tiempo_operacion': np.random.normal(8, 1, n),
    'numero_operarios': np.random.randint(1, 5, n),
    'calidad_materia': np.random.normal(80, 10, n)
})
X_prod['rendimiento'] = (0.6 * X_prod['velocidad_maquina'] +
                         0.3 * X_prod['calidad_materia'] +
                         5 * X_prod['numero_operarios'] +
                         np.random.normal(0, 5, n))

# Dividir datos y entrenar regresión lineal
y = X_prod.pop('rendimiento')
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(X_prod, y, test_size=0.3, random_state=42)

reg = LinearRegression()
reg.fit(X_train_p, y_train_p)
y_pred = reg.predict(X_test_p)
mse = mean_squared_error(y_test_p, y_pred)
print(f"Error cuadrático medio (MSE): {mse:.2f}")
```

```python
# Visualizar peso de cada variable en la predicción
coefs = pd.Series(reg.coef_, index=X_prod.columns)
coefs.sort_values().plot(kind='barh')
plt.title("Coeficientes del modelo de regresión")
plt.xlabel("Peso en la predicción del rendimiento")
plt.show()
```
